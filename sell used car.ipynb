# converting categorical column to numerical column

```python
df_c['Mileage'] = pd.to_numeric (df_c['Mileage'].str.split().str.get(0),errors = 'coerce')
```

```python
df_c.head()
```

```python
df_c['Engine'] = pd.to_numeric (df_c['Engine'].str.split().str.get(0),errors = 'coerce')
```

```python
df_c.head()
```

```python
df_c['Power'] = pd.to_numeric (df_c['Power'].str.split().str.get(0),errors = 'coerce')
```

```python
df_c.head()
```

```python
df_c.isnull().sum()
```

```python
df_c.isnull().mean()*100#  if any column the  missing value is more than 56 then drop it
                        #  if any column the missing value is less than 56 then it replace with the median value
                        #  if any column the missing value is 1 - 2 percent then remove that column
```

```python
df_c.drop('New_Price',axis = 1,inplace = True)
```

```python
df_c['Engine'] = df_c['Engine'].fillna(value = df_c['Engine'].median())
```

```python
df_c['Power'] = df_c['Power'].fillna(value = df_c['Power'].median())
```

```python
df_c['Seats'] = df_c['Seats'].fillna(value = df_c['Seats'].median())
```

```python
df_c['Price'] = df_c['Price'].fillna(value = df_c['Price'].median())
df_c.dropna(axis = 0,inplace =True)
```
